```python
import cv2
import numpy as np

low_yellow = (20, 100, 100)
high_yellow = (30,255,255)
kernal_er=np.ones((2,2),np.uint8)
```

```python
count=0
pi_cnt=0
prv_pi_cnt=0
count_list=[]
cap=cv2.VideoCapture("bottlecounts.mp4")
while(cap.isOpened()):
    ret, frame=cap.read()
    if not ret:
        break
        
    #print(frame.shape) 
    #frame_blur=cv2.blur(frame,(5,5))
    
    cv2.rectangle(frame,(195,320),(250,365),(0,255,0),3)
    im_cut=frame[320:365,195:250]
    #blur_cut=cv2.blur(im_cut,(3,3))
    #hsv_cut=cv2.cvtColor(blur_cut,cv2.COLOR_BGR2HSV)
    hsv_cut=cv2.cvtColor(im_cut,cv2.COLOR_BGR2HSV)
    mask_yellow = cv2.inRange(hsv_cut, low_yellow, high_yellow)
    er_im=cv2.erode(mask_yellow,kernal_er,iterations=3)
    print(cv2.countNonZero(er_im))
    pi_cnt=cv2.countNonZero(er_im)
    count_list.append(pi_cnt)
    
    if (pi_cnt <40) &(prv_pi_cnt>40):
        count=count+1
        print("the bottle count is",count)
        
    frame=cv2.putText(frame, "Bottles : ", (190, 310),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 55, 255), 2)
    frame=cv2.putText(frame, str(count), (260, 310),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 55, 255), 2)     
    
    cv2.imshow("frame1",mask_yellow)
    #cv2.imshow("frame",er_im)
    cv2.imshow("colour",frame)
    prv_pi_cnt=pi_cnt
    if cv2.waitKey(100) & 0xFF==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
